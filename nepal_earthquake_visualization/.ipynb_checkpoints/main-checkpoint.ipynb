# Nepal Earthquake Visualization

This notebook visualizes earthquake data for Nepal using the USGS Earthquake API.

```python
# Install required packages if running in a new environment
# !pip install pandas requests matplotlib folium
```

```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from datetime import datetime, timedelta, timezone
```

## Fetch Earthquake Data
Fetching earthquake data for Nepal from the USGS API (last 365 days).

```python
end_date = datetime.now(timezone.utc)
start_date = end_date - timedelta(days=365)

# Nepal bounding box
min_lat, max_lat = 26.347, 30.446
min_lon, max_lon = 80.058, 88.201

url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
params = {
    'format': 'geojson',
    'starttime': start_date.strftime('%Y-%m-%d'),
    'endtime': end_date.strftime('%Y-%m-%d'),
    'minlatitude': min_lat,
    'maxlatitude': max_lat,
    'minlongitude': min_lon,
    'maxlongitude': max_lon,
    'minmagnitude': 2.5,
    'limit': 1000
}

response = requests.get(url, params=params)
if response.status_code == 200:
    data = response.json()
    #print(f"API Response Status Code: {response.status_code}")
    #print(f"Sample of API Response: {str(data)[:200]}...") # Print a sample of the response
else:
    print(f"Error fetching data: {response.status_code}")
    data = {'features': []} # Ensure data is an empty dictionary with 'features' key

earthquakes = []
if 'features' in data and len(data['features']) > 0:
    for feature in data['features']:
        prop = feature['properties']
        geom = feature['geometry']
        earthquakes.append({
            'place': prop.get('place', 'Unknown'),
            'mag': prop.get('mag', None),
            'time': pd.to_datetime(prop['time'], unit='ms'),
            'lat': geom['coordinates'][1],
            'lon': geom['coordinates'][0]
        })
else:
    print('No earthquake data found for Nepal in the last year.')

df = pd.DataFrame(earthquakes)
df.head()
```

## Magnitude Distribution
Histogram of earthquake magnitudes.

```python
if not df.empty:
    plt.figure(figsize=(10,6))
    plt.hist(df['mag'], bins=20, color='skyblue', edgecolor='black')
    plt.title('Earthquake Magnitude Distribution in Nepal (Last 365 Days)')
    plt.xlabel('Magnitude')
    plt.ylabel('Count')
    plt.show()
else:
    print('No data to plot magnitude distribution.')
```

## Interactive Map
Map of earthquake locations and heatmap.

```python
if not df.empty:
    center_lat = df['lat'].mean()
    center_lon = df['lon'].mean()
    m = folium.Map(location=[center_lat, center_lon], zoom_start=7)
    # Add markers
    for _, row in df.iterrows():
        folium.CircleMarker([row['lat'], row['lon']],
            radius=max(row['mag']*1.5, 2),
            color='red',
            fill=True,
            fill_opacity=0.7,
            popup=f"{row['place']} (Mag: {row['mag']})"
        ).add_to(m)
    # Add heatmap
    HeatMap(df[['lat', 'lon']].values, radius=15).add_to(m)
    m
else:
    print('No data to plot map.')
```

## Summary of Analysis

```python
if not df.empty:
    largest_eq = df.loc[df['mag'].idxmax()]
    summary = f'''
    ### Nepal Earthquake Analysis Summary

    This analysis visualized earthquake data for Nepal from the USGS API (last 365 days).

    **Key Findings:**
    - Total earthquakes analyzed: {len(df)}
    - Largest earthquake: {largest_eq['mag']} magnitude near {largest_eq['place']} on {largest_eq['time'].strftime('%Y-%m-%d')}
    - Earthquakes are distributed across Nepal, with higher frequency in certain regions.

    **Recommendations:**
    - Strengthen earthquake preparedness and monitoring in high-frequency regions.
    '''
    print(summary)
else:
    print('No data available for summary.')
```
